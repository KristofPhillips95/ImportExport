```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_prices = pd.read_csv("../Results/TradeCurves/import_price_curves2025_2012_National Trends_8736_s_false_tc_BE00_TYNDP_gs_BE00_DE00_FR00_UK00_NL00_100.csv")
```

```python
filtered_columns = [col for col in df_prices.columns if float(col) % 2000 == 0]
```

```python
df_prices[filtered_columns].boxplot()
```

```python
df_filtered = df_prices[filtered_columns]
mean_values = df_filtered.mean()

# Create a bar plot for mean values
plt.figure(figsize=(10, 6))
mean_values.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Mean Prices by Import Level')
plt.xlabel('Import Levels')
plt.ylabel('Mean Prices')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```
