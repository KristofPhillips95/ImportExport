```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Loop through each DataFrame and plot boxplots for each column
for i, (name, df) in enumerate(dfs_prices.items()):
    # Convert column names to float and filter to retain only multiples of 1000
    filtered_columns = [col for col in df.columns if (float(col) % 500 == 0) & (abs(float(col)) <= 10000)]
    df_filtered = df[filtered_columns]
    
    # Create boxplot for each column
    positions = [p + i * 0.2 for p in range(len(filtered_columns))]
    df_filtered.boxplot(positions=positions, widths=0.2, patch_artist=True, boxprops=dict(facecolor='skyblue'), showmeans=True, meanprops=dict(marker='o', markerfacecolor='red', markersize=8), meanline=True, label=name)

plt.title('Mean Prices by Import Level')
plt.xlabel('Import Levels')
plt.ylabel('Prices')
plt.xticks(ticks=[p + 0.4 for p in range(len(filtered_columns))], labels=filtered_columns, rotation=45)  # Set x-axis ticks and labels
plt.legend()
plt.tight_layout()
plt.show()
```
