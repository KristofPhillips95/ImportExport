```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_model_type_name(row,columns):
    strings = [str(row[col]) for col in columns if pd.notna(row[col])]
    return '-'.join(strings)
```

```python
# df_results_1 = pd.read_csv("../Results/InvestmentModelResults/Loop_4models_neighbors_only_LUGFR_out_8760.csv")
df_results_2 = pd.read_csv("../Results/InvestmentModelResults/Loop_4models_indirect_neighbors_8760.csv")
df_results_3 = pd.read_csv("../Results/InvestmentModelResults/Loop_4models_indirect_neighbors_storage_8760.csv")
df_results_4 = pd.read_csv("../Results/InvestmentModelResults/Loop_4models_granularity_8760.csv")
```

```python
df_results = pd.concat([df_results_2,df_results_3,df_results_4])
#df_results = pd.read_csv("../Results/InvestmentModelResults/Loop_4models_granularity_8760.csv")
```

```python
geo_scope_map = {
    'A_l_l': 'All',
    'DE00_NL00_FR00_UK00_BE00_LUG1': 'DN',
    'N_A': ''
}
df_results["geoscope_m"] = df_results["geoscope"].replace(geo_scope_map)

# List of columns to consider for joining
columns_to_join = ['type', 'trans_cap_other', 'geoscope_m',"stepsize","target_cap_for_curves","simplified"]

# Apply the function row-wise to create the new column
df_results['name'] = df_results.apply(create_model_type_name, args=(columns_to_join,), axis=1)

name_map = {
    'NTC-1.0e10-DN': 'SN-DN',
    'NTC-1.0e10-All': 'SN-All',
    'NTC-S-All': "NTC-All",
    'NTC-S-DN': "NTC-DN",
    'TCPC-All': "TCPC-All",
    'TCPC-DN': "TCPC",
    
}
df_results['Model Type'] = df_results["name"].replace(name_map)
```

```python
# Get deviations from benchmark model
bm_t = "NTC-S-All-False"

# Filter the DataFrame for the benchmark model
col_bm = df_results[(df_results["Model Type"] == bm_t)]

# Initialize an empty list to store deviations
dev_data = []

vals_dev = ["CCGT", "OCGT", "PV", "w_off", "w_on","investment_cost"]

# Calculate deviations from the benchmark model
for index, row in df_results.iterrows():
    t = row["type"]
    stepsize = row["stepsize"]
    trans_cap_other = row["trans_cap_other"]
    tcfc = row["target_cap_for_curves"]
    simplified = row["simplified"]
    geo_scope = row["geoscope_m"]
    model_type = row["Model Type"]
    
    
    deviations = row[vals_dev] - col_bm.iloc[0][vals_dev]# Accessing the first row of col_bm for deviations
    perc_deviations = (row[vals_dev] - col_bm.iloc[0][vals_dev])/col_bm.iloc[0][vals_dev]
    weighted_abs_deviations = abs(row[vals_dev] - col_bm.iloc[0][vals_dev])*col_bm.iloc[0][vals_dev]
    
    # Create a dictionary for the deviations
    dev_dict = {"Model Type":model_type,"type": t, "stepsize": stepsize, "trans_cap_other": trans_cap_other,"target_cap_for_curves":tcfc,"simplified":simplified,"geo_scope":geo_scope}
    for col, dev,per_dev,weighted_abs_deviation in zip(vals_dev, deviations,perc_deviations,weighted_abs_deviations):
        dev_dict[col] = per_dev
        dev_dict[col+"_w"] = weighted_abs_deviation
        #dev_dict[col+"_dev"] = per_dev
    
    # Append the dictionary to the list
    dev_data.append(dev_dict)

# Create DataFrame from the list of dictionaries
dev_df = pd.DataFrame(dev_data)
dev_df["Mean deviation"] = dev_df[vals_dev].abs().mean(axis=1)
```

```python
#filter = (dev_df["type"] == "TCS") & (dev_df["simplified"] == False)
filter =  (dev_df["simplified"] == False)#& (dev_df["geo_scope"] == "All")
```
